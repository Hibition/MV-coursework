```python
from PIL import Image
import os

target_shape = (256, 256)

healthData_path =
cancerData_path = r'D:\Project\Deep Learning - Research\programs\MV-couswork\Brain Tumor Data Set\BrainTumorProcessed\Brain Tumor'
dataset_path = cancerData_path

hd_dis = r'D:\Project\Deep Learning - Research\programs\MV-couswork\Brain Tumor Data Set\BrainTumorProcessed\Healthy'
cd_dis = r'D:\Project\Deep Learning - Research\programs\MV-couswork\Brain Tumor Data Set\BrainTumorProcessed\Brain Tumor'

image_files = [f for f in os.listdir(dataset_path) if f.endswith('.jpg') or f.endswith('.png')]

i = 1
for image_file in image_files:
    image_path = os.path.join(dataset_path, image_file)
    after_path = os.path.join(cd_dis, image_file)
    
    if i % 100 == 0:
        print(i)
    i = i + 1
    
    img = Image.open(image_path)
    scale_factor = max(target_shape[0] / img.size[0], target_shape[1] / img.size[1])

    img_resized = img.resize((int(img.size[0] * scale_factor), int(img.size[1] * scale_factor)), Image.ANTIALIAS)
    
    center_x, center_y = img_resized.size[0] // 2, img_resized.size[1] // 2

    crop_left = max(center_x - target_shape[0] // 2, 0)
    crop_top = max(center_y - target_shape[1] // 2, 0)
    crop_right = min(center_x + target_shape[0] // 2, img_resized.size[0])
    crop_bottom = min(center_y + target_shape[1] // 2, img_resized.size[1])
    
    img_cropped = img_resized.crop((crop_left, crop_top, crop_right, crop_bottom))
    img_cropped = img_cropped.convert('RGB')
    img_cropped.save(after_path)

print("Image reshape finished: ", target_shape)
```

```python
import os
import random
import shutil

# root path
dataset_root = r'D:\Data\jupyter notebook\MV-couswork\Brain Tumor Data Set\BrainTumorProcessed'

cancer_dir = cd_dis
healthy_dir = hd_dis

# decide precentage
train_ratio = 0.7
val_ratio = 0.15
test_ratio = 0.15

train_dir = os.path.join(dataset_root, "train")
val_dir = os.path.join(dataset_root, "val")
test_dir = os.path.join(dataset_root, "test")

# Create target path
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Get image path
cancer_images = [f for f in os.listdir(cancer_dir) if f.endswith('.jpg') or f.endswith('.png')]
healthy_images = [f for f in os.listdir(healthy_dir) if f.endswith('.jpg') or f.endswith('.png')]

# reorder the dataset
random.shuffle(cancer_images)
random.shuffle(healthy_images)

cancer_train_idx = int(len(cancer_images) * train_ratio)
cancer_val_idx = int(len(cancer_images) * (train_ratio + val_ratio))

healthy_train_idx = int(len(healthy_images) * train_ratio)
healthy_val_idx = int(len(healthy_images) * (train_ratio + val_ratio))

cancer_train_images = cancer_images[:cancer_train_idx]
cancer_val_images = cancer_images[cancer_train_idx:cancer_val_idx]
cancer_test_images = cancer_images[cancer_val_idx:]

healthy_train_images = healthy_images[:healthy_train_idx]
healthy_val_images = healthy_images[healthy_train_idx:healthy_val_idx]
healthy_test_images = healthy_images[healthy_val_idx:]

for img in cancer_train_images:
    shutil.copy(os.path.join(cancer_dir, img), os.path.join(train_dir, img))
for img in cancer_val_images:
    shutil.copy(os.path.join(cancer_dir, img), os.path.join(val_dir, img))
for img in cancer_test_images:
    shutil.copy(os.path.join(cancer_dir, img), os.path.join(test_dir, img))
for img in healthy_train_images:
    shutil.copy(os.path.join(healthy_dir, img), os.path.join(train_dir, img))
for img in healthy_val_images:
    shutil.copy(os.path.join(healthy_dir, img), os.path.join(val_dir, img))
for img in healthy_test_images:
    shutil.copy(os.path.join(healthy_dir, img), os.path.join(test_dir, img))

print("Data split finish.")
```
